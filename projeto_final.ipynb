Importação dos módulos

```python
import pandas as pd
import numpy as np
from unidecode import unidecode
```

Importação dos dataframes

```python
df_escolas = pd.read_csv('escolas.csv')
df_material_didatico = pd.read_csv('material_didatico.csv')
df_subprefeituras = pd.read_csv('subprefeituras.csv')
df_dados_entrega = df_escolas.copy()
pd.reset_option('display.max_columns')
pd.reset_option('display.max_rows')
# df_dados_entrega
```

Nome das colunas em snake_case

```python
df_material_didatico.rename(columns=lambda x: x.lower(), inplace=True)
df_subprefeituras.rename(columns=lambda x: x.lower(), inplace=True)
df_dados_entrega.rename(columns=lambda x: x.lower(), inplace=True)
# df_dados_entrega
```

Retirada dos acentos

```python

df_dados_entrega['escolas_postos'] = df_dados_entrega['escolas_postos'].apply(unidecode)
df_dados_entrega['bairro'] = df_dados_entrega['bairro'].apply(unidecode)
df_dados_entrega['endereço '] = df_dados_entrega['endereço '].apply(unidecode)
df_subprefeituras['nome'] = df_subprefeituras['nome'].apply(unidecode)
df_subprefeituras['subprefeitura'] = df_subprefeituras['subprefeitura'].apply(unidecode)
# df_dados_entrega


```

Remoção das escolas com id duplicado


```python
df_dados_entrega = df_dados_entrega.drop_duplicates(subset=['id'])
df_material_didatico = df_material_didatico.drop_duplicates(subset=['id'])
# df_dados_entrega = df_dados_entrega.drop_duplicates()
# df_material_didatico = df_material_didatico.drop_duplicates()
# df_dados_entrega['id'].value_counts()
# pd.unique(df_dados_entrega['id'])
# df_dados_entrega 
```

Todas as strings em maiúsculo

```python

df_dados_entrega.loc[:, 'endereço '] = df_dados_entrega['endereço '].str.upper()
df_subprefeituras.loc[:, 'nome'] = df_subprefeituras['nome'].str.upper()
df_subprefeituras.loc[:, 'subprefeitura'] = df_subprefeituras['subprefeitura'].str.upper()
# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
# df_dados_entrega
```

Padronização dos nomes


```python
df_dados_entrega.loc[:, 'endereço '] = df_dados_entrega['endereço '].str.upper()
df_dados_entrega.loc[:, 'endereço '] = df_dados_entrega['endereço '].str.replace('AV.','AVENIDA')
df_dados_entrega.loc[:, 'endereço '] = df_dados_entrega['endereço '].str.replace('AV ','AVENIDA')
df_dados_entrega.loc[:, 'endereço '] = df_dados_entrega['endereço '].str.replace('R.','RUA')
df_dados_entrega.loc[:, 'endereço '] = df_dados_entrega['endereço '].str.replace('PCA.','PRACA')
df_dados_entrega.loc[:, 'endereço '] = df_dados_entrega['endereço '].str.replace('ESTR.','ESTRADA')
# df_dados_entrega.loc[:, 'endereço '] = df_dados_entrega['endereço '].str.replace(' O','')
df_dados_entrega.loc[:, 'endereço '] = df_dados_entrega['endereço '].str.replace('O.',' ')
df_dados_entrega.loc[:, 'endereço '] = df_dados_entrega['endereço '].str.replace('N.','N')
df_dados_entrega.loc[:, 'endereço '] = df_dados_entrega['endereço '].str.replace('.','')
df_dados_entrega.loc[:, 'endereço '] = df_dados_entrega['endereço '].str.replace('DEG','')
df_dados_entrega.loc[:, 'endereço '] = df_dados_entrega['endereço '].str.replace(',','')
df_dados_entrega.loc[:, 'endereço '] = df_dados_entrega['endereço '].str.replace('-','')



# df_dados_entrega
```

Separando logradouro do número, exclusão da coluna endereço e tratando NaN

```python


df_dados_entrega.loc[:,'logradouro'] = df_dados_entrega['endereço '].str.replace(r'(\d+|S/NO)', '', regex=True)
df_dados_entrega.loc[:,'numero'] = df_dados_entrega['endereço '].str.extract(r'(\d+|S/N)')
df_dados_entrega.loc[:,'numero'] = df_dados_entrega.fillna('NAO INFORMADO')

df_dados_entrega = df_dados_entrega.drop('endereço ', axis=1)
df_dados_entrega




```

Verificando parte inicial dos nomes das Escolas para Criação do TIPO de escola

```python
somente_tipos = df_dados_entrega['escolas_postos'].str.split(' ', n=1).str.get(0)

somente_tipos_unicos = somente_tipos.unique()

print(somente_tipos_unicos)
```

Filtro para tipos de escolas (Escola Municipal, Ciep e Colegio Municipal)

```python
condicao_em_1 = df_dados_entrega['escolas_postos'].str.contains('EM')
condicao_em_2 = df_dados_entrega['escolas_postos'].str.contains('ESCOLA MUNICIPAL')
condicao_em_3 = df_dados_entrega['escolas_postos'].str.contains('E.M')
condicao_ciep_1 = df_dados_entrega['escolas_postos'].str.contains('CENTRO INTEGRADO DE EDUCACAO PUBLICA')
condicao_ciep_2 = df_dados_entrega['escolas_postos'].str.contains('CIEP')
condicao_cm = df_dados_entrega['escolas_postos'].str.contains('COLEGIO MUNICIPAL')

condicao_escolas_municipais = df_dados_entrega[condicao_em_1 | condicao_em_2 | condicao_em_3]
condicao_ciep = df_dados_entrega[condicao_ciep_1 | condicao_ciep_2]
condicao_colegio = df_dados_entrega[condicao_cm]




```

Criando coluna 'tipo' com base no tipo de escola (Ciep, Escola Municipal ou Colegio Municipal)

```python
def tipo_de_escola(escola):
    if escola in condicao_escolas_municipais['escolas_postos'].values:
        return 'ESCOLA'
    elif escola in condicao_ciep['escolas_postos'].values:
        return 'CIEP'
    elif escola in condicao_colegio['escolas_postos'].values:
        return 'COLEGIO'
    else:
        return 'NAO INFORMADO'
    


for escola in df_dados_entrega['escolas_postos'].values:
    df_dados_entrega['tipo'] = df_dados_entrega['escolas_postos'].apply(tipo_de_escola)
df_dados_entrega
```

Correção dos ids para 3 caracteres

```python
df_dados_entrega['id'] = df_dados_entrega['id'].astype(str).str.zfill(3)

df_dados_entrega
```

Criação da coluna subprefeitura



```python
def tipo_de_subprefeitura(bairro):
    for nome, subprefeitura in df_subprefeituras.values:
        if bairro == nome.upper():
            return subprefeitura.upper()
        elif bairro == 'MARACANA/ VILA ISABEL' or bairro == 'MARACANA/ TIJUCA':
            return 'GRANDE TIJUCA'
        elif bairro == 'RECREIO':
            return 'BARRA DA TIJUCA'
        elif bairro == 'OSWALDO CRUZ':
            return 'ZONA NORTE'
        elif bairro == 'FREGUESIA JPA' or bairro == 'NOVA SEPETIBA' or bairro == 'RIO DAS PEDRAS':
            return 'JACAREPAGUA'          
        elif bairro == 'AUGUSTO VASCONCELOS':
            return 'ZONA OESTE'    
    return 'NAO INFORMADA'
 
for bairro in df_dados_entrega['bairro'].values:
    df_dados_entrega['subprefeitura'] = df_dados_entrega['bairro'].apply(tipo_de_subprefeitura)

# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
df_dados_entrega
```

Correção do nome das colunas latitude e longitude e mudança para 5 casas decimais

```python
def correcao_lat_e_lon(numero):
    numero_com_ponto = numero.replace(',', '.')
    numero_float = float(numero_com_ponto)
    numero_arredondado = round(numero_float,5)
    return numero_arredondado 

df_dados_entrega['lat'] = df_dados_entrega['lat'].apply(correcao_lat_e_lon)
df_dados_entrega['lon'] = df_dados_entrega['lon'].apply(correcao_lat_e_lon)
df_dados_entrega = df_dados_entrega.rename(columns={'lat': 'latitude','lon':'longitude'})
df_dados_entrega
```

Criação da coluna 'quantidade de material' 


```python

df_dados_entrega['id'] = df_dados_entrega['id'].astype(int)
df_dados_entrega = pd.merge(df_dados_entrega,df_material_didatico, on='id')
df_dados_entrega = df_dados_entrega.rename(columns={'quantidade':'quantidade_de_material'})

df_dados_entrega.loc[:,'quantidade_de_material'] = df_dados_entrega.fillna(0)
df_dados_entrega.loc[:,'quantidade_de_material'] = df_dados_entrega['quantidade_de_material'].replace("''", 0)
df_dados_entrega.loc[:,'quantidade_de_material'] = df_dados_entrega['quantidade_de_material'].replace("'NaN'", 0)
df_dados_entrega.loc[:,'quantidade_de_material'] = df_dados_entrega['quantidade_de_material'].replace("' '", 0)
df_dados_entrega.loc[:,'quantidade_de_material'] = df_dados_entrega['quantidade_de_material'].replace("'","")
df_dados_entrega.loc[:,'quantidade_de_material'] = df_dados_entrega['quantidade_de_material'].replace("'397'","397")
df_dados_entrega.loc[:,'quantidade_de_material'] = df_dados_entrega['quantidade_de_material'].replace("'395'","395")

df_dados_entrega['quantidade_de_material'] = df_dados_entrega['quantidade_de_material'].astype(int)



df_dados_entrega

```

Reorganizando as colunas


```python

df_dados_entrega = df_dados_entrega[['id','escolas_postos','tipo','logradouro','numero','bairro','subprefeitura','latitude','longitude','quantidade_de_material']]
df_dados_entrega = df_dados_entrega.rename(columns = {'escolas_postos':'nome'})
df_dados_entrega['id'] = df_dados_entrega['id'].astype(str).str.zfill(3)
# df_dados_entrega.set_index('id',inplace=True)
df_dados_entrega
```

Total de material por subprefeitura

```python

material_por_subprefeitura = df_dados_entrega.groupby('subprefeitura')['quantidade_de_material'].sum()
df_material_por_subprefeitura = pd.DataFrame(material_por_subprefeitura)
df_material_por_subprefeitura.reset_index(inplace= True)
df_material_por_subprefeitura.to_csv('material_por_subprefeitura.csv', sep=',', index=False)
df_material_por_subprefeitura

```
